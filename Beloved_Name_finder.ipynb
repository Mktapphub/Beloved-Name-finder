```python
import turtle
import random
import string
import math
import time


def draw_heart():
    """Draws a beautiful red heart."""
    heart.penup()
    heart.goto(0, -200)  # Position heart at the center of the screen
    heart.pendown()
    heart.color("red")
    heart.fillcolor("red")
    heart.begin_fill()
    heart.left(140)
    heart.forward(180)
    heart.circle(-90, 200)
    heart.setheading(60)
    heart.circle(-90, 200)
    heart.forward(180)
    heart.end_fill()
    heart.penup()
    heart.home()


def display_text(letter):
    """Displays the given letter inside the heart."""
    text.clear()
    text.color("white")
    text.goto(0, -100)  # Adjusted position for better centering
    text.write(
        letter,
        align="center",
        font=("Arial", 30, "bold"),  # Reduced font size
    )


def animate_letter_in_heart(user_name):
    """Splashes random letters, reveals the final letter, and exits after a delay."""
    splashed_letters = []

    for _ in range(10):
        random_letter = random.choice(string.ascii_uppercase)
        splashed_letters.append(random_letter)

        text.clear()
        text.write(random_letter, align="center", font=("Arial", 30, "bold"))
        time.sleep(0.2)

    # Select the final letter based on user_name
    if user_name.lower() == "minhajul":
        final_letter = "A"  # Set final letter to 'A' for Minhajul
    else:
        final_letter = random.choice(splashed_letters)  # Random for others

    # Display the final letter
    display_text(final_letter)

    text.goto(0, 200)
    text.write(
        "We found your beloved!",
        align="center",
        font=("Courier", 16, "italic"),
    )
    time.sleep(5)  # Reduced delay before exit

    turtle.bye()


def splash_letter_inside_heart(x, y):
    """Refreshes the heart and triggers the letter animation."""
    heart.clear()  # Clear previous heart
    draw_heart()  # Draw a new heart


# Setup Turtle Screen
screen = turtle.Screen()
screen.title("Beloved name Finder")
screen.bgcolor("black")
screen.setup(width=800, height=600)

# Create Turtle for heart
heart = turtle.Turtle()
heart.speed(3)
heart.hideturtle()

# Create Turtle for text
text = turtle.Turtle()
text.hideturtle()
text.penup()

# Draw initial heart and display instructions
draw_heart()
text.goto(0, 200)
text.color("white")
text.write(
    "Tap anywhere to unveil the letter of your beloved!",
    align="center",
    font=("Courier", 16, "italic"),
)

# Get  name before starting the animation
user_name = turtle.textinput("Enter Your Name", "Please enter your name:")

# Click to trigger the animation and reveal the letter
screen.onclick(lambda x, y: [splash_letter_inside_heart(x, y), animate_letter_in_heart(user_name)])
screen.mainloop()
```
